```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#load the dataset
epl_df = pd.read_csv ("C:\\Users\\HP\\Videos\\Research videos\\MS excel\\EPL_20_21.csv")
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
#adding the minspermatch and goals per match column
epl_df['MinsPerMatch'] = (epl_df ['Mins']/epl_df ['Matches']).astype (int)
epl_df ['GoalsPerMatch'] = (epl_df ['Goals']/epl_df ['Matches']).astype (float)
print (epl_df.head())
```

```python
#total goals
total_goals = epl_df ['Goals'].sum()
print (total_goals)
```

```python
#total penalty goals
total_penalty_goals= epl_df ["Penalty_Goals"].sum()
total_penalty_goals
```

```python
#total penalty attempts
total_penalty_attempts = epl_df ["Penalty_Attempted"].sum()
total_penalty_attempts
```

```python
#piechart for penalties missed vs scored
plt.figure(figsize = (13,6))
plt_not_scored = epl_df ['Penalty_Attempted'].sum() - total_penalty_goals
data = [plt_not_scored, total_penalty_goals]
labels = ['Penalties missed', 'Penalties scored']
plt.pie (data, labels= labels, autopct= '%.0f%%')
plt.show ()
```

```python
#unique positions
epl_df['Position'].unique()
```

```python
#players from different nations
np.size((epl_df['Nationality'].unique()))
```

```python
#most players from which countries
nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot (kind ='bar', figsize=(12,6))   
```

```python
#clubs with most players in their squad
epl_df ['Club'].value_counts().nlargest(6).plot(kind = 'bar')
```

```python
#clubs with least players in their squad
epl_df["Club"].value_counts().nsmallest(8).plot(kind = "bar")
```

```python
under_20 = epl_df[epl_df['Age']<= 20]
age20_25 = epl_df[(epl_df['Age']> 20) & (epl_df ["Age"]<= 25)]
age25_30 = epl_df[(epl_df['Age']> 25) & (epl_df ["Age"]<= 30)]
Above_30 = epl_df[epl_df['Age']> 30]
```

```python
#pie chart for players between different age limits
x = np.array ([under_20 ['Name'].count(),age20_25["Name"].count(),age25_30['Name'].count(),Above_30['Name'].count()])
mylabels = ['<=20', '>20 & <=25' , '.25 & <=30', '>30']
plt.title ("Total Players with Age Groups", fontsize = 25)
plt.pie (x,labels = mylabels, autopct= '%.1f%%')
plt.show
```

```python
#players under the age of 20
player_under_20 = epl_df [epl_df ["Age"] < 20]
player_under_20 ["Club"].value_counts().plot (kind = 'bar')
```

```python
#average age of players in each club
plt.figure(figsize= (12, 5))
sns.boxplot(x = 'Club', y = 'Age', data = epl_df)
plt.xticks (rotation = 90)
```

```python
num_player = epl_df.groupby('Club').size()
data = epl_df.groupby ('Club')['Age'].sum()/num_player
data.sort_values(ascending=False)
```

```python
#total assists by club
Total_assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index= False)['Assists'].sum())
sns.set_theme(style= 'whitegrid', color_codes=True)
ax = sns.barplot(x="Club", y = 'Assists', data=Total_assists_by_clubs.sort_values (by="Assists"))
ax.set_xlabel ("Club", fontsize = 30)
ax.set_ylabel ("Assists", fontsize = 24)
plt.xticks(rotation = 75)
plt.rcParams ["figure.figsize"] = (20, 10)
plt.title ("Plot of Clubs  vs Total Assists", fontsize  = 25)
```

```python
#top 10 assists
top_10_assists= epl_df [["Name", "Club", "Assists" ,"Matches"]].nlargest(n=10, columns= "Assists")
top_10_assists
```

```python
#goals by clubs
Goals_from_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style= 'whitegrid', color_codes=True)
ax = sns.barplot(x="Club", y = 'Goals', data=Goals_from_clubs.sort_values (by="Goals"), palette= 'rocket')
ax.set_xlabel ("Club", fontsize = 30)
ax.set_ylabel ("Goals", fontsize = 24)
plt.xticks(rotation = 75)
plt.rcParams ["figure.figsize"] = (20, 10)
plt.title ("Plot of Clubs  vs Total Goals", fontsize  = 25)

```

```python
#top 10 goals
top_10_goals= epl_df [["Name", "Club", "Goals" ,"Matches"]].nlargest(n=10, columns= "Goals")
top_10_goals
```

```python
#goals per match
goals_per_match= epl_df [["Name", "GoalsPerMatch", "Matches", "Goals"]].nlargest(n=10, columns= "GoalsPerMatch")
goals_per_match
```

```python
#pie chart to show the goals without assists vs goals with assists
plt.figure (figsize= (15, 8))
assists = epl_df ['Assists'].sum()
data = [total_goals - assists, assists]
labels=['Goals without assists', 'Goals with assists']
plt.pie (data, labels = labels, autopct= '%.0f%%')
plt.show()
```

```python
#players with the most red cards
epl_red = epl_df.sort_values(by= 'Red_Cards', ascending=False)[:10]
plt.figure(figsize=(22, 10))
plt.title ('players with the most red cards')
c=sns.barplot(x=epl_red ['Name'], y= epl_red['Red_Cards'], label = 'Players', color= 'red')
```
